```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
import pandas as pd

# Select predictors
df = pd.read_csv("d:\DTD\Project\Dataset\Data transformation.csv")
X = df[['R&D_Spend','Marketing_Spend','Administration']] 
X_const = sm.add_constant(X)

# Compute VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i+1) 
                   for i in range(len(X.columns))]
print(vif_data)


# merge R&D_Spend,Marketing_Spend
X['Growth_Spend'] = X['R&D_Spend'] + X['Marketing_Spend']
#X = X.drop(columns=['R&D_Spend','Marketing_Spend'])
print(X.head())
# Scale the data
SKLEARN_AVAILABLE = True
if SKLEARN_AVAILABLE:
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X = pd.DataFrame(X_scaled, columns=X.columns)
else:
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)
    X = pd.DataFrame(X_scaled, columns=X.columns)
#add the Growth_Spend scaled column in the dataset


# Download the new dataset
X.to_csv("d:\DTD\Project\Dataset\Data transformation_vif1.csv", index=False)
```
